# torchvision 이란?
Torchvision은 PyTorch와 함께 사용되는 Computer vision용 라이브러리이다. 효율적인 이미지 및 비디오 변환을 위한 유틸리티, 일반적으로 사용되는 사전 학습된 모델 및 데이터 세트가 몇가지 있다.


# torchvision의 장점
- PyTorch와 함께 제공되기 때문에 자동으로 GPU 지원이 함께 제공된다. (속도가 빠르다)
- 구현이 간단하다. 
- 일반적으로 사용되는 사전 훈련된 모델과 샘플 스타터 코드들이 제공한다.
- Facebook AI팀에서 개발 및 유지 관리하며 Python 커뮤니티에서 지원한다.

```python
import torch
import torchvision
from torchvision import transforms
import torchvision.datasets
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# 두개의 이미지 비교 출력 함수
row, col = 50, 50
def compare_two_img(img1, img2):
    plt.figure(figsize=(row, col))
    plt.subplot(1, 2, 1)
    plt.imshow(img1)
    plt.subplot(1, 2, 2)
    plt.imshow(img2)
    
```

```python
my_transform = transforms.Compose([
    transforms.RandomGrayscale(p=1),       # 그레이 스케일 (흑백으로 변환)
    transforms.CenterCrop(100),              # CenterCrop (중앙에서 100x100 픽셀 이미지로 이미지 자르기)
    transforms.RandomHorizontalFlip(p=0.5),  # HorizontalFlip (수평으로 뒤집기)
    transforms.ToTensor(),                   # ToTensor (모든 것을 이미지 텐서로 변환)
])
```

```python
my_dataset = torchvision.datasets.CIFAR10(root='./data', transform=my_transform, download=True)
my_dataloader = torch.utils.data.DataLoader(my_dataset, batch_size=16, shuffle=True, num_workers=4)

plt.imshow(my_dataset[3][0].permute(1,2,0))
```

```python
my_transform = transforms.Compose([
    transforms.ColorJitter(brightness=(0.2, 2), 
                               contrast=(0.3, 2), 
                               saturation=(0.2, 2), 
                               hue=(-0.3, 0.3))
])

my_dataset = torchvision.datasets.CIFAR10(root='./data', download=True)
my_dataset_trans = torchvision.datasets.CIFAR10(root='./data', transform=my_transform, download=True)

compare_two_img(my_dataset[4][0], my_dataset_trans[4][0])
```

```python
my_transform = transforms.Compose([
    transforms.GaussianBlur(19)
])

my_dataset = torchvision.datasets.CIFAR10(root='./data', download=True)
my_dataset_trans = torchvision.datasets.CIFAR10(root='./data', transform=my_transform, download=True)

compare_two_img(my_dataset[4][0], my_dataset_trans[4][0])
```

```python
my_transform = transforms.Compose([
    transforms.Pad(5)
])

my_dataset = torchvision.datasets.CIFAR10(root='./data', download=True)
my_dataset_trans = torchvision.datasets.CIFAR10(root='./data', transform=my_transform, download=True)

compare_two_img(my_dataset[4][0], my_dataset_trans[4][0])
```

```python
my_transform = transforms.Compose([
    transforms.RandomAffine(90)
])

my_dataset = torchvision.datasets.CIFAR10(root='./data', download=True)
my_dataset_trans = torchvision.datasets.CIFAR10(root='./data', transform=my_transform, download=True)

compare_two_img(my_dataset[4][0], my_dataset_trans[4][0])
```
